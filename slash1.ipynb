```python
import pandas as pd

# Try reading the CSV with different encodings
encodings = ['utf-8', 'latin1', 'ISO-8859-1', 'cp1252']
for enc in encodings:
    try:
        df = pd.read_csv('Amazon Sale Report.csv', encoding=enc)
        print(f'Successfully read the file with encoding: {enc}')
        break
    except UnicodeDecodeError as e:
        print(f'Failed to read the file with encoding: {enc} - {e}')
```

```python

# Inspect the first few rows of the dataset
print(df.head())

# Check the data types of each column
print(df.info())
```

```python
# Generate summary statistics for numerical and categorical variables
print(df.describe(include='all'))

# Visualize the distribution of key features to identify trends and patterns
import matplotlib.pyplot as plt
import seaborn as sns

# Identify numerical and categorical columns
numerical_columns = df.select_dtypes(include=['float64', 'int64']).columns
categorical_columns = df.select_dtypes(include=['object']).columns

# Example: Distribution of all numerical columns
for col in numerical_columns:
    sns.histplot(df[col].dropna(), bins=30)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()

# Example: Count plot for all categorical columns
for col in categorical_columns:
    sns.countplot(y=col, data=df)
    plt.title(f'Count of {col}')
    plt.xlabel('Count')
    plt.ylabel(col)
    plt.show()

```
